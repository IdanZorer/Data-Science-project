```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"../df_games.csv", header=0, sep=',')
```

```python
df['price'] = np.round(df['price'],3)

fig = plt.figure(figsize=(5, 5))
cat_plot = df.groupby('price')['installs'].mean()
plt.xticks(cat_plot.tolist())
plt.ylabel('Installs')
cat_plot.plot(kind='line')
```

```python
fig = plt.figure(figsize=(5, 5))
cat_plot = df.groupby('installs')['rating_count'].mean()
plt.ylabel('Rating Count')
plt.title('Installs')
cat_plot.plot(kind='line')
```

```python
df['rating'] = np.round(df['rating'],1)

fig = plt.figure(figsize=(5, 5))
plot = df.groupby('rating')['installs'].mean()
plt.ylabel('Installs')
plot.plot(kind='line')
```

```python
fig = plt.figure(figsize=(5, 5))
cat_plot = df.groupby('categories')['installs'].sum()
plt.yticks(cat_plot.values.tolist())
plt.ylabel('Downloads')
plt.title('Downloads Per Categories')
cat_plot.plot(kind='bar')
```

```python
fig = plt.figure(figsize=(5, 5))
cat_plot = df.groupby('categories')['price'].mean()
plt.ylabel('price')
cat_plot.plot(kind='bar')
```
